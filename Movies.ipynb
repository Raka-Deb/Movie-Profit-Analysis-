```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv("/Users/rakadeb/Desktop/Data_Analysis/movie_profit.csv")

# Display first five rows
print(data.head())
```

```python
# Display last five rows
print(data.tail())
```

```python
# Display first three rows
print(data.head(3))
```

```python
# Check column names
print(data.columns)
```

```python
# Drop the 'Unnamed: 0' column
data.drop(columns=['Unnamed: 0'], inplace=True)
print(data.head())
```

```python
# Get general information about the dataset
print(data.info())
```

```python
# Check for missing values
print(data.isna().sum())
```

```python
# Drop rows with missing values
data.dropna(inplace=True)
print(data.isna().sum())
```

```python
# Check for duplicates
print(data.duplicated().sum())
```

```python
# Get descriptive statistics for numerical columns
print(data.describe())
```

```python
# Sort by production budget (ascending)
data.sort_values(by='production_budget', ascending=True, inplace=True)
print(data.head())
```

```python
# Value counts for genre
print(data['genre'].value_counts())
```

```python
# Group by rating and calculate mean domestic gross
print(data.groupby('mpaa_rating')['domestic_gross'].mean().sort_values(ascending=False))
```

```python
# Group by rating and get top two
print(data.groupby('mpaa_rating')['domestic_gross'].mean().head(2))
```

```python
# Histogram of worldwide gross
plt.hist(data['worldwide_gross'])
plt.title('Distribution of Worldwide Gross')
plt.xlabel('Worldwide Gross')
plt.ylabel('Frequency')
plt.show()
```

```python
# Bar plot of release date vs domestic gross (first few rows)
plt.bar(data['release_date'].head(), data['domestic_gross'].head())
plt.xlabel('Date')
plt.ylabel('Domestic Gross')
plt.show()
```

```python
# Group by distributor and calculate sum of worldwide gross
print(data.groupby('distributor')['worldwide_gross'].sum())
```

```python
# Corrected group by distributor and sum of worldwide gross
print(data.groupby('distributor')['worldwide_gross'].sum().sort_values(ascending=False).head(2))
```

```python
# Group by genre and calculate mean production budget
print(data.groupby('genre')['production_budget'].mean())
```

```python
# Group by rating and calculate median domestic gross
print(data.groupby('mpaa_rating')['domestic_gross'].median())
```

```python
# Group by year (from release_date) and count movies
print(data.groupby(data['release_date'].str.slice(0, 4))['movie'].count().sort_values(ascending=False).head())
```

```python
# Group by genre and find maximum domestic gross
print(data.groupby('genre')['domestic_gross'].max())
```

```python
# Group by rating and find minimum production budget
print(data.groupby('mpaa_rating')['production_budget'].min())
```

```python
# Group by distributor and calculate average worldwide gross
print(data.groupby('distributor')['worldwide_gross'].mean())
```

```python
# Group by rating and calculate standard deviation of worldwide gross
print(data.groupby('mpaa_rating')['worldwide_gross'].std())
```

```python
# Bar chart of genre vs sum of production budget
data.groupby('genre')['production_budget'].sum().plot(kind='bar')
plt.title('Total Production Budget by Genre')
plt.xlabel('Genre')
plt.ylabel('Total Production Budget')
plt.show()
```

```python
# Line chart of year vs mean worldwide gross
data.groupby(data['release_date'].str.slice(0, 4))['worldwide_gross'].mean().plot(kind='line')
plt.title('Mean Worldwide Gross Over Years')
plt.xlabel('Release Year')
plt.ylabel('Mean Worldwide Gross')
plt.show()
```

```python
# Pie chart of rating vs sum of domestic gross
data.groupby('mpaa_rating')['domestic_gross'].sum().plot(kind='pie')
plt.title('Total Domestic Gross by Rating')
plt.ylabel('')
plt.show()
```

```python
# Scatter plot of production budget vs worldwide gross
data.plot(kind='scatter', x='production_budget', y='worldwide_gross')
plt.title('Scatter Plot of Production Budget vs Worldwide Gross')
plt.xlabel('Production Budget')
plt.ylabel('Worldwide Gross')
plt.show()
```

```python
# Horizontal bar chart of distributor vs mean domestic gross (top few)
data.groupby('distributor')['domestic_gross'].mean().head().plot(kind='barh')
plt.title('Average Domestic Gross by Distributor')
plt.xlabel('Average Domestic Gross')
plt.ylabel('Distributor')
plt.show()
```
